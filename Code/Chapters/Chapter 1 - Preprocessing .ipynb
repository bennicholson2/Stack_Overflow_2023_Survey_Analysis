### 1.1 Data handling: Columns with multiple responses
The following columns need to be modified for analysis
- MainBranch Categorical answer with 6 different responses that can be numbered 1-6
- Employment Categorical answer that has option for multiple responses for one column which requires further normalisation
- RemoteWork Categorical answer with 3 different responses that can be numbered 1-3
- Coding Activities Categorical as well as nominal, this has an open ended question which is going to require NLP work to create a word count frequency.

```python
#Create function that is used to get the different responses and create a numbered response that will be used to facilitate the new dataframe

def get_unique_response_values(df, column_list): 
    """
    Parameters:
    df is the dataframe that the unique values are coming from
    column_list is the list of columns that you would like to go through and get unique responses for
    """
    unique_values_dict = {}  # Dictionary to hold unique values for each column
    for col in column_list:
        unique_values_dict[col] = df[col].unique()  # Store unique values for each column
    
    # Find the maximum number of unique values among all columns
    max_unique_values = max(len(unique_values_dict[col]) for col in column_list)
    
    # Create an empty DataFrame with columns as specified in column_list
    empty_df = pd.DataFrame(columns=column_list)
    
    # Fill DataFrame with unique values
    for i in range(max_unique_values):
        row_data = {}
        for col in column_list:
            unique_values = unique_values_dict[col]
            if i < len(unique_values):
                row_data[col] = unique_values[i]
            else:
                row_data[col] = pd.NA  # Fill with NaN if uneven number of unique values
        empty_df = empty_df.append(row_data, ignore_index=True)
    
    return empty_df
```

### Problem with function
Notice how columns 'MainBranch' and 'RemoteWork' have returned reasonable length responses, however 'Employment' had the option has multiple responses for each column and 'CodingActivties' allowed for open ended responses. 

#### 1.1.1 Employment Column

Employment is made up of 9 different options, however multiple answers can be chosen. The problem with this is that there are 9! ways that a user could have responded to this question. To overcome this issue, there a few methods.
1. Duplicate responses

This would result in the duplication of data for each different response. For example if the respondent had 'Employed, full time' and 'Independent contractor' this would result in two new columns. This would be effective in getting a count of the type of employment however, this creates data dependencies. If a lot of the higher income earners are doing multiple jobs at once then it would result in it seeming that the average income would be higher than what it is. As keeping the integrity of the SO data survey is integral to analysis another handling technique is going to be required.

2. Dummy Variables

Using a dummy variable would create a new column that has a binary indicator of the variables for each option. This is one that might have practical issues. However you could keep the integrity of the other data points and be able to get a count of the different types of employment. This is going to work better for the data. 

A word count is not necessary as it was not an open response to the question. So creating, 9 new dummy variables where they are as the following
- Employment, full time
- Employed, part-time
- Independent contractor, freelancer, or self-employed
- Not employed, but looking for work
- Not employed, and not looking for work
- Student, full-time
- Student, part-time
- Retired
- I prefer not to say

```python
import pandas as pd

def preprocess_df_column(dataframe, column_name):
    """
    Preprocesses the 'Employment' column in the given DataFrame by replacing commas,
    creating dummy variables, concatenating them with the original DataFrame, and filling NaN values with 0.
    
    Parameters:
    dataframe (DataFrame): The DataFrame containing the 'Employment' column.
    column_name (str): The name of the column to preprocess.
    
    Returns:
    DataFrame: The updated DataFrame with preprocessed 'Employment' column.
    """
    # Copy the DataFrame to avoid modifying the original
    df = dataframe.copy()
    
    # Replace commas in the 'Employment' column
    df[column_name] = df[column_name].astype(str).str.replace(',', '')

    # Create dummy variables
    dummy_df = df[column_name].str.get_dummies(sep=';')

    # Concatenate the dummy variables with the original DataFrame
    df = pd.concat([df, dummy_df], axis=1)

    # Fill NaN values with 0
    df = df.fillna(0)

    return df
```

```python
basic_info_df_with_employment_dummies = preprocess_df_column(basic_info_df,'Employment')
basic_info_df_with_employment_dummies.drop(columns=['I prefer not to say','nan'],inplace=True)
basic_info_df_with_employment_dummies
```

The function above has created a dataframe that includes dummy variables and has successfully create a binary response for the type of employment a user has. This can now be used in better analysis.
